```
sings = pd.read_csv('data/signamesad.csv')
```

```
sings.head()
```

```
#slownik pomocniczy
sings.to_dict()['b']
```

```
dict_sings = sings.to_dict()['b']
```

```
"""funkcja pomocnicza"""

for id_sing in dict_sings.keys():

  given_signe = X_train[y_train == id_sing]

  plt.figure(figsize=(15,5))
  for i in range(9):
    plt.subplot('19{0}'.format(i+1))
    plt.imshow(given_signe[0])
    plt.axis('off')

plt.tight_layout
plt.show()
```

```
#kontener ile razy wystapuje znak o danym 
#id znaku
#ile razy wystapil
Counter(y_train).most_common()
```

```
cnt = Counter(y_train).most_common()

#rozbija na liste id i liste wystapien
id_lables, cnt_labes = zip(*cnt)

#id_lables, cnt_labes

ids = range(len(id_lables))

plt.figure(figsize=(15,5))
plt.bar(ids,cnt_labes)
plt.xlabel('Znaki')
plt.title('Znaki drogowe - wg liczby wysapie')
 
## id_ z petli id labels
## id_lables[id_] - id z barplota,
## dict_sings [id_lables[id_] - id ze slownika pomocnicznego
lables =  [dict_sings [id_lables[id_]] for id_ in id_lables ] 
plt.xticks(ids, lables,rotation = 'vertical')

plt.show()
```
